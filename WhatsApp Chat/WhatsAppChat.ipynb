```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
# Define the pattern to match timestamps and messages
# pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s(.*?)\n'
# Extract dates and messages separately
# matches = re.findall(pattern, data)
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
import pandas as pd

# Your code to extract messages and dates

df = pd.DataFrame({'user-message': messages, 'message-date': dates})
df['message-date'] = pd.to_datetime(df['message-date'], format='%d/%m/%y, %I:%M\u202f%p - ')
df.rename(columns={'message-date': 'date'}, inplace=True)
df.sample(10)
```

```python
df.shape
```

```python
users = []
messages = []

for message in df['user-message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[0])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user-message'], inplace=True)
df.sample()
```
